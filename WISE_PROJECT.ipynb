```python
df = pd.get_dummies(df)
```

```python
# Using dummies can increase the volume of data - called dummy variable trap
```

```python
df = df.drop(columns=['GENDER_M','LUNG_CANCER_NO'])
df.head()
```

```python
predictor = df.drop(columns=['LUNG_CANCER_YES'])
target = df['LUNG_CANCER_YES']
```

```python
predictor.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
predictor = ss.fit_transform(predictor)
```

```python
# predictor.head() - This will not execute because it is in array
```

```python
predictor[5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(predictor,target,test_size=0.2,random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)

lr_predict = lr.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(lr_predict,y_test))
```

```python
# SVM(KERNEL = 'RBF/LINEAR)
# NAIVE BAYES
# KNN
# DECISION TREE(CRITERION = 'GINI/ENTROPY')
# RANDOM FOREST(CRITERION = 'GINI/ENTROPY)
```

```python
# SVM
from sklearn.svm import SVC

svm = SVC(kernel='linear')

svm.fit(x_train,y_train)

svm_predicted = svm.predict(x_test)

print(accuracy_score(svm_predicted,y_test))
```

```python
#rbf kernel
from sklearn.svm import SVC

svm = SVC(kernel='rbf')

svm.fit(x_train,y_train)

svm_predicted = svm.predict(x_test)

print(accuracy_score(svm_predicted,y_test))
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train,y_train)
gnb_predicted = gnb.predict(x_test)

print(accuracy_score(gnb_predicted,y_test))
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)

neigh.fit(x_train,y_train)
neigh_predicted = neigh.predict(x_test)

print(accuracy_score(neigh_predicted,y_test))
```

```python
#decision tree

from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='gini')
clf = clf.fit(x_train,y_train)

clf_predicted = clf.predict(x_test)

print(accuracy_score(clf_predicted,y_test))
```

```python
clf_e = tree.DecisionTreeClassifier(criterion='entropy')
clf_e = clf_e.fit(x_train,y_train)

clfe_predicted = clf.predict(x_test)

print(accuracy_score(clfe_predicted,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion='gini')
rfc = rfc.fit(x_train,y_train)

rfc_predicted = rfc.predict(x_test)

print(accuracy_score(rfc_predicted,y_test))

```

```python
rfc_e = RandomForestClassifier(criterion='entropy')
rfc_e = rfc_e.fit(x_train,y_train)

rfce_predicted = rfc_e.predict(x_test)

print(accuracy_score(rfce_predicted,y_test))
```

```python
from time import time
parameters = {'penalty' : ['l1', 'l2', 'elasticnet', None],
              'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
              'max_iter' : [100, 200, 300, 400, 500],
              }
#we are checking in the different variation of numerical values
#in the parameters of random forest classifier

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = lr, # Make sure classifier points to the RF model
                           param_grid = parameters,
                           scoring = "accuracy",
                           cv = 10,#from k-fold validation
                           n_jobs = -1)#to use the entire core of the machine

t0 = time()#just to check how much time it takes to excute the program
grid_search = grid_search.fit(x_train, y_train)
t1 = time()
print("Took %0.2f seconds" % (t1 - t0))

rf_best_accuracy = grid_search.best_score_
rf_best_parameters = grid_search.best_params_
rf_best_accuracy, rf_best_parameters
```
